# Sinusoids

### Formula for a CT Sinusoid:

### $$x(t) = A\cos(\omega t + \phi)$$

- $A$ : amplitude
- $\omega$ : radian frequency
- $\phi$ : phase

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-2,2,.01)
omega_1 = np.pi
x =  np.cos(omega_1 * t)
plt.plot(t,x)
plt.title("$f(t) = \cos(\omega_1 t)$")
plt.show()
plt.plot(np.cos(omega_1 *1.5 * t))
plt.title("$g(t) = \cos(1.5 \omega_1 t )$")
_=plt.plot()
```

## Two signals:

### $$x_1 = \cos(\omega_1 t + a)$$
### $$x_2 = \cos(\omega_1 t - a)$$



```python
a = np.pi / 4
x1 = np.cos(omega_1 * t + a)
x2 = np.cos(omega_1 * t - a)
x0 = np.cos(omega_1 * t)
plt.plot(t,x1)
plt.title("$x_1$")
plt.show()
plt.plot(t,x2)
plt.title("$x_2$")
plt.show()
```

If $a$ is positive, the signal is shifted left. If it is negative, the signal is shifted right.

## $$\omega \text{ vs } f$$

$$\omega : \frac{rad}{sec}$$


$$f : \frac{cycles}{second}$$


$$\omega = 2\pi f$$

# Discrete Frequency

- For the graph of $\cos(\Omega t)$ below,it may be true that $\Omega = \frac{\pi}{5}$.

- The continuous graph of $\cos(\frac{\pi}{5} t)$ fits the points perfectly

```python
t = np.arange(0,15,1)
Omega = np.pi / 5
x = np.cos(Omega*t)
plt.stem(t,x)
plt.title("$x[t]=\cos(\Omega t)$")
t2 = np.arange(0,15,.01)
x1 = np.cos(Omega*t2)
plt.show()
plt.plot(t2,x1,color='black')
plt.stem(t,x)
plt.show()
```

# Other possible solutions

- The points may have been sampled at a different rate

```python
t = np.arange(0,15,1)
Omega = np.pi / 5
x = np.cos(Omega*t)
plt.stem(t,x)
plt.title("$x[t]=\cos(\Omega t)$")
t2 = np.arange(0,15,.01)
x1 = np.cos(2.2*np.pi*t2)
plt.show()
plt.plot(t2,x1,color='black')
plt.stem(t,x,)
plt.show()
```

# Aliasing

- As we only hae values at integer multiples of $\Omega$, there are_multiple_ possible values of $\Omega$ that lead to the same set of points
- This is called _aliasing_, the same signal can be described by different names/aliases

## Base Band

- The frequency $0 \leq \Omega \leq \pi$ (in the __base band__) is the _principle alias_

## Practice Problem

$$x[n] = \cos \left(\frac{5\pi}{4}\right)$$

How can we represent this with a frequency $\Omega$ in the base band?

$$\frac{5\pi}{4} - 2\pi = \frac{-3\pi}{4}$$


$$\cos\left(\frac{-3\pi}{4}\right) = \cos\left(\frac{3\pi}{4}\right)$$

as $\cos$ is an __even__ signal.

# Maximum Frequency

The maximum possible discrete frequency in the base band is:

## $$\Omega_{\text{max}} = \pi$$


For continuous signals, there is no maximum frequency.

# Sampling

- __Sampling__ is used to convert CT signals to DT (discretizing)
- $T$ (seconds/sample) = sampling interval
- $f_s$ (samples/second) = sampling rate
- How does sampling affect the information in the signal?

# Sampling Loss

- No information between samples is known
- Information is distorted

# In class Example:

- Playing a song at increasingly lower sampling rates
- Starting at $f_s$ = 44 kHz, ending at 344 Hz
- Progressively sounds more and more distorted
- Interesting note: base is not distorted as much as the rest of the sound
- Higher frequencies alias down to lower frequencies, causing lower distortions


## Efects of samplin

- NyQuist frequency
- If frequencies in the CT signal are greater than the ase bad, they alias to frequencies in the base band
- _Anti-aliasing_ can be used to prevent distortions
- Remove frequencies above Nyquist before sampling
