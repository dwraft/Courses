# Signals

- CT (continuous) vs DT (discrete)
- Physical signals are often CT such as time or spatial coordinates
- Computations manipulate functions of discrete domain (time samples, spatial coordinate samples

## Periodic

- Periodic signals consist of repeated cycles (sine waves, sawtooth)
- Aperiodic do not repeat

## Left vs Right 

- Right sidesd signals are zero before some start time
- Left sided signals are zero after some ending time


## Bounded Signal

- The minima and maxima of a bounded signal are finite

## Even vs Odd

- Odd signal (symmetric about $t=0$), $x(a)=x(-a)$
- Odd signals are antisymmetric about $t=0$, $x(a) = -x(-a)$

# Complex Numbers

- We use $j$...
- $a + bj$ is _rectangular form_
- Other way to think about it: like the vector $<a,b>$ in $\mathbb{R}^2$
- Alternate representation
- $r$ and $\theta$
- $\theta = \arctan(\frac{b}{a})$
- $r = \|<a,b>\| = \sqrt{a^2+b^2}$
- Note: $\theta$ may be called _angle_ or _phase_ and occasionaly _argument_

## Euler's Formula


$$c = r(\cos(\theta) + j\sin(\theta)$$
$$ c = re^{j\theta}$$

Both forms are useful for different appliciations

## Addition

- Rectangular form is convenient

$$c_1 + c_2 = a_1 + b_1 j + a_2 + b_2 j$$
$$=(a_1+a_2) + (b_1 + b_2) j$$

## Multiplication

- Euler form is convenient

$$c_1 \cdot c_2 = r_1 r_2 e ^ {j(\theta_1 + \theta_2)}$$
